```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('/Users/siddharthtripathi/Downloads/Pandas/student-scores.csv')

X = data[['weekly_self_study_hours']]
Y = data['math_score']
model = LinearRegression()
model.fit(X,Y)
predicted_scores = model.predict(X)

MAE = mean_absolute_error(Y, predicted_scores)
MSE = mean_squared_error(Y, predicted_scores)
RMSE = np.sqrt(MSE)
R2 = r2_score(Y,predicted_scores)

print(f"Mean Absolute Error: {round(MAE,2)}")
print(f"Mean Squared Error: {round(MSE,2)}")
print(f"Root Mean Squared Error: {round(RMSE,2)}")
print(f"R-squared: {round(R2,2)}")

plt.figure(figsize=(10,6))
plt.hist(Y, bins=20, alpha=1, color='coral', edgecolor='black')
plt.hist(predicted_scores, bins=20, alpha=1, color='skyblue', edgecolor='black')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Actual vs Predicted Math Scores")
plt.xlabel("Math Scores")
plt.ylabel("Number of Students")
plt.legend(["Actual Scores", "Predicted Scores"])
plt.tight_layout()
plt.savefig('Actual_vs_Predicted_Math_Scores.png', dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(10,6))
plt.scatter(X,Y, color='red', edgecolor='black')
plt.plot(X, predicted_scores, color='blue', linewidth=2)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Weekly Self Study Hours vs Math Scores")
plt.xlabel("Weekly Self Study Hours")
plt.ylabel("Math Scores")
plt.tight_layout()
plt.savefig('Weekly_Self_Study_Hours_vs_Math_Scores.png', dpi=300, bbox_inches='tight')
plt.show()

new_hours = 10
predicted_new_score = model.predict([[new_hours]])
print(f"Predicted math score for a student studying {new_hours} hours per week: {round(predicted_new_score[0],2)}")
```
